#### 4. Відсутні дані
Видно, що відсутні дані позначаються знаком "?". Визначемо в який стовпчиках існують такі значення.

```python
dataFile.isin(['?']).any()
```

Отримаємо кількість відсутніх значень в цих стовпчиках.

```python
print('normalized-losses', len(dataFile[dataFile['normalized-losses'] == '?']))
print('num-of-doors', len(dataFile[dataFile['num-of-doors'] == '?']))
print('bore', len(dataFile[dataFile['bore'] == '?']))
print('stroke', len(dataFile[dataFile['stroke'] == '?']))
print('horsepower', len(dataFile[dataFile['horsepower'] == '?']))
print('peak-rpm', len(dataFile[dataFile['peak-rpm'] == '?']))
print('price', len(dataFile[dataFile['price'] == '?']))
```

Враховуючи результати, бачимо, що стовпчик 'normalized-losses' має суттєві втрати значень в порівнянні з іншими. Тож при аналізі варто задуматися щодо його використання.
